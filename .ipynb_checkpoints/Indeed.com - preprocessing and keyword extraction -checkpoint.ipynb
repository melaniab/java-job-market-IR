```python
#Tokenizing
from nltk.tokenize import TweetTokenizer
tweet_tok = TweetTokenizer()
tokens = tweet_tok.tokenize(full_text)
print("All tokens: ", len(tokens))
```

```python
print(tokens[0:20])
```

```python
tokens_lower = [token.lower() for token in tokens]
print(tokens_lower[0:20])
```

```python
#Remove punctuation 
tokens_no_punct = [token for token in tokens_lower if token not in ".!?-/\()[]{},:;`'\""]
#remove one-words tokens 
token_words = [token for token in tokens_no_punct if len(token) > 1]
len(token_words)
```

```python
#Removing stopwords 
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
filtered_words = [word for word in token_words if word not in stopwords.words('english')]
print("With no stopwords: ", len(filtered_words))
print(filtered_words[0:20])
```

```python
unique_lemmas = list(set(lemmas))
print("Unique words: ", len(unique_lemmas))
```

```python
#Finding new words 
from nltk.corpus import words
nltk.download('words')
candidates = [lemma for lemma in unique_lemmas if lemma not in words.words()]
```
